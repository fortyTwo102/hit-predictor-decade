```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import numpy as np

dataset = pd.read_csv(os.path.join('dataset', 'dataset-of-10s.csv'))
data = dataset.iloc[:, 3:]

combo = [('loudness', 'energy'), ('loudness', 'danceability'), ('acousticness', 'energy'), \ 
          ('valence', 'energy'), ('valence', 'danceability'), ('valence', 'loudness'), \
        ('instrumentalness', 'acousticness'), ('sections', 'durations'), ('danceability', 'speechiness')]

```

```python
sns.set(style="white")
# Generate a large random dataset
d = data
# Compute the correlation matrix
corr = d.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
